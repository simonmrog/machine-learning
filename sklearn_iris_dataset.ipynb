```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot") #changin plot style to dots
iris = datasets.load_iris() #using the dataset iris from sklearn package

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

print ("Iris dataset key names: ")
print(iris.keys()) #labels of the iris dataset
print ("Iris dataset feature names: ")
print(iris.feature_names) #features of the iris dataset
print ("Iris dataset target names: ")
print(iris.target_names) #labels of the target variable: three species of iris

X = iris.data #creating a numpy array with the features data
y = iris.target #creating a numpy array with the target data

df = pd.DataFrame(X, columns = iris.feature_names) #creating the data frame

df.head() #shows the first five observations
pd.scatter_matrix(df, c = y, figsize = [8,8], s = 150, marker = "D") #creates the plot
```
